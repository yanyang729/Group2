#  Prepared data

```python
import pandas as pd
import itertools
from collections import Counter
```

```python
df = pd.read_csv('./preprocesse_samples_from_first_100000.csv',sep=',',quotechar='|',header=None)
df.columns = ['title','body','tags']
```

```python
df.head(3)
```

```python
for i in range(1,100,10):
    print(df.body[i],'\n')
```

```python
all_tags = df.tags.apply(lambda x: x.replace('<','').split('>'))
all_tags = [x for x in list(itertools.chain(*all_tags)) if x ]
```

```python
ct = Counter(all_tags)
vocab = ct.most_common()
```
